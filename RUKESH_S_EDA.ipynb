```python
merged_data = df_3.merge(df_1, on='CustomerID').merge(df_2, on='ProductID')
merged_data
```

```python
top_selling_products = merged_data.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(5)
print("Top 5 Products Sold:")
print(top_selling_products)
```

```python
plt.figure(figsize=(10, 6))
top_selling_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
category_product_count = df_2['Category'].value_counts().sort_values(ascending=False)
print(category_product_count)

```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Price', data=df_2)
plt.title('Price Range for Each Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```

```python
top_customer_transactions = merged_data.groupby('CustomerName')['TransactionID'].count().sort_values(ascending=False).head(1)
print("\nCustomer with the most transactions:")
print(top_customer_transactions)
```

```python
most_sold_product = merged_data.groupby('ProductName')['Quantity'].sum().idxmax()
print("\nProduct that sold the most:", most_sold_product)
```
